```python
### Figure 2 ###

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import csv

# Use Viridis palette
# High‐contrast Viridis shades:
viridis = [
    "#440154",  # deep purple
    "#3b528b",  # rich blue
    "#21918c",  # teal
    "#5ec962",  # lighter green
]



def read_nonzero_column(path, column):
    vals = []
    with open(path, 'r') as f:
        reader = csv.DictReader(f)
        for row in reader:
            try:
                v = float(row[column])
                if v > 0:
                    vals.append(v)
            except:
                continue
    return np.array(vals)

def compute_cdf(vals, divide_by_1000=False):
    s = np.sort(vals) / (1000.0 if divide_by_1000 else 1.0)
    y = np.arange(1, s.size+1) / s.size
    return s, y

# load the four series
dw2_ms = read_nonzero_column('figure_3/cab_dwaas_2.csv',   'query_duration')
dw4_ms = read_nonzero_column('figure_3/cab_dwaas_4.csv',   'query_duration')
dw6_ms = read_nonzero_column('figure_3/cab_dwaas_6.csv',   'query_duration')
qa_s   = read_nonzero_column('figure_3/cab_qaas.csv', 'query_duration')

# compute CDFs
x2, y2 = compute_cdf(dw2_ms, divide_by_1000=True)
x4, y4 = compute_cdf(dw4_ms, divide_by_1000=True)
x6, y6 = compute_cdf(dw6_ms, divide_by_1000=True)
xq, yq = compute_cdf(qa_s,  divide_by_1000=False)

# 80th-percentile cutoff
p80s = [np.percentile(arr, 80) for arr in (x2, x4, x6, xq)]
cutoff = max(p80s) * 1.05

# Figure 1: Bulk (≤80th pct)
fig1, ax1 = plt.subplots(figsize=(5, 3.6), dpi=120)
for data, color, label, ls in [
    ((x2, y2), viridis[0], 'DWaaS N=2', ':'),
    ((x4, y4), viridis[1], 'DWaaS N=4', '-.'),
    ((x6, y6), viridis[2], 'DWaaS N=6', '--'),
    ((xq, yq), viridis[3], 'QaaS',      '-')
]:
    xs, ys = data
    ax1.plot(xs, ys, label=label,
             color=color, linestyle=ls, linewidth=3)

ax1.set_xlim(0, cutoff)
ax1.set_xlabel("Query Latency (sec)", fontsize=22)
ax1.set_ylabel("CDF", fontsize=22)
ax1.tick_params(labelsize=21, width=1.5, length=6)
ax1.grid(True, linestyle='--', linewidth=0.5, alpha=1.0)

# L-shaped black spines
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
for side in ('bottom','left'):
    ax1.spines[side].set_linewidth(1.5)
    ax1.spines[side].set_color('black')

ax1.legend(
    frameon=False,
    fontsize=22,
    loc='lower right',
    bbox_to_anchor=(1, -0.05),
    labelspacing=0.1,
    handlelength=1.0,    # shorter line in legend
    markerscale=0.7,     # smaller marker in legend
    handletextpad=0.5,
    borderpad=0.3
)

fig1.tight_layout()
fig1.savefig("2_bulk_cdf.pdf", format="pdf", bbox_inches="tight", pad_inches=0.1)


# Figure 2: Full Tail (0–987s)
fig2, ax2 = plt.subplots(figsize=(5, 3.6), dpi=120)
for data, color, label, ls in [
    ((x2, y2), viridis[0], 'DWaaS N=2', ':'),
    ((x4, y4), viridis[1], 'DWaaS N=4', '-.'),
    ((x6, y6), viridis[2], 'DWaaS N=6', '--'),
    ((xq, yq), viridis[3], 'QaaS',      '-')
]:
    xs, ys = data
    ax2.plot(xs, ys, label=label,
             color=color, linestyle=ls, linewidth=3)

ax2.set_xlim(0, 987)
ax2.set_xlabel("Query Latency (sec)", fontsize=22)
ax2.set_ylabel("CDF", fontsize=22)
ax2.tick_params(labelsize=21, width=1.5, length=6)
ax2.grid(True, linestyle='--', linewidth=0.5, alpha=1.0)

ax2.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)
for side in ('bottom','left'):
    ax2.spines[side].set_linewidth(1.5)
    ax2.spines[side].set_color('black')

ax2.legend(
    frameon=False,
    fontsize=22,
    loc='lower right',
    bbox_to_anchor=(1, -0.05),   # shift legend 5% of axes height below the axes
    labelspacing=0.1,
    handletextpad=0.5,
    borderpad=0.3
)

fig2.tight_layout()
fig2.savefig("2_tail_cdf.pdf", format="pdf", bbox_inches="tight", pad_inches=0.1)

plt.show()
```

```python
### Figure 3 ###

import os
import csv
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
import seaborn as sns

# ----------------------------------------------------------------
# --- Minimal throughput computation (active throughput only) ----
# ----------------------------------------------------------------

def compute_active_throughput(csv_path):
    """Compute active throughput (queries per active second) from a CSV file."""
    starts, durations = [], []

    with open(csv_path, "r") as f:
        reader = csv.DictReader(f)
        for row in reader:
            try:
                s = float(row["start"])
                d = float(row["query_duration_with_queue"])
                if s >= 0 and d > 0:
                    starts.append(s)
                    durations.append(d)
            except (ValueError, KeyError):
                continue

    if not starts:
        print(f"⚠️ No valid rows in {csv_path}")
        return 0.0

    starts = np.array(starts)
    durations = np.array(durations)

    # Handle special case: QaaS durations are in seconds, not ms
    if "qaas" in csv_path.lower():
        durations *= 1000.0

    # Compute active time (union of running intervals)
    intervals = sorted(zip(starts, starts + durations), key=lambda x: x[0])
    merged = []
    for s, e in intervals:
        if not merged or s > merged[-1][1]:
            merged.append([s, e])
        else:
            merged[-1][1] = max(merged[-1][1], e)
    active_time_s = sum(e - s for s, e in merged) / 1000.0

    n_queries = len(starts)
    throughput_active = n_queries / active_time_s if active_time_s > 0 else 0.0

    print(f"📄 {csv_path}")
    print(f"   Queries: {n_queries}")
    print(f"   Active time: {active_time_s:.2f} s")
    print(f"   Throughput (active): {throughput_active:.3f} q/s\n")

    return throughput_active


# ----------------------------------------------------------------
# --- Compute throughputs for both workloads ---------------------
# ----------------------------------------------------------------

def gather_throughputs(base_dir="figure_4"):
    """
    Compute active throughputs for:
      - steady/dwaas.csv
      - steady/ep.csv
      - bursty/dwaas.csv
      - bursty/ep.csv
    Returns two lists: perf_constant, perf_spiky
    """
    paths = {
        "steady": {
            "DWaaS": os.path.join(base_dir, "steady", "dwaas.csv"),
            "EP": os.path.join(base_dir, "steady", "ep.csv"),
        },
        "bursty": {
            "DWaaS": os.path.join(base_dir, "bursty", "dwaas.csv"),
            "EP": os.path.join(base_dir, "bursty", "ep.csv"),
        },
    }

    # Compute throughputs dynamically
    dwaas_steady = compute_active_throughput(paths["steady"]["DWaaS"])
    ep_steady = compute_active_throughput(paths["steady"]["EP"])
    dwaas_spiky = compute_active_throughput(paths["bursty"]["DWaaS"])
    ep_spiky = compute_active_throughput(paths["bursty"]["EP"])

    # Fixed QaaS throughput (same for both workloads)
    qaas_throughput = 1.59

    # Lists correspond to [DWaaS, EP, QaaS]
    perf_constant = [dwaas_steady, ep_steady, qaas_throughput]
    perf_spiky = [dwaas_spiky, ep_spiky, qaas_throughput]

    return perf_constant, perf_spiky


# ----------------------------------------------------------------
# --- Figure 3 plotting ------------------------------------------
# ----------------------------------------------------------------

def plot_figure3(perf_constant, perf_spiky):
    sns.set_theme(style='whitegrid', context='talk', font_scale=1.8)

    labelz = ['DWaaS', 'EP', 'QaaS']
    colors = sns.color_palette("viridis", 3)

    # Monetary costs — keep fixed as in original
    cost_constant = [7.32, 11.39, 16.78]
    cost_spiky = [8.78, 5.88, 4.31]

    arrow_props = dict(fc='gray', ec='gray', headwidth=15, width=2, headlength=12)

    fig, ax = plt.subplots(figsize=(12,4))

    # Annotate “better”
    ax.annotate(
        'better',
        xy=(0.05, 2), xycoords=('axes fraction','data'),
        xytext=(5, 1.72), textcoords='data',
        ha='center', fontsize=26, fontweight='bold',
        arrowprops=arrow_props,
        clip_on=False
    )

    # Plot lines
    ax.plot(cost_constant, perf_constant,
            'o--', color='black', linewidth=4, markersize=12,
            markerfacecolor='white', markeredgecolor='black',
            label='Constant Workload')
    ax.plot(cost_spiky, perf_spiky,
            's:', color='black', linewidth=4, markersize=12,
            markerfacecolor='white', markeredgecolor='black',
            label='Spiky Workload')

    # Scatter points
    ax.scatter(cost_constant, perf_constant,
               color=colors, s=200, marker='o',
               edgecolor='black', linewidth=2, zorder=5)
    ax.scatter(cost_spiky, perf_spiky,
               color=colors, s=200, marker='s',
               edgecolor='black', linewidth=2, zorder=5)

    ax.set_xlim(0, 22)
    ax.set_ylim(0, 2)

    ax.set_xlabel('Monetary Cost ($)', fontsize=26)
    ax.set_ylabel('Compl.\n Queries/s', fontsize=26)

    ax.tick_params(axis='both', labelsize=26, width=2, length=8)
    ax.grid(True, linestyle='--', linewidth=1.0, alpha=1.0)
    ax.set_axisbelow(True)

    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    for side in ['bottom','left']:
        ax.spines[side].set_color('black')
        ax.spines[side].set_linewidth(3)

    legend1 = ax.legend(
        loc='lower left',
        bbox_to_anchor=(0, -0.12),
        fontsize=26,
        frameon=False,
        handlelength=1.5,
        handletextpad=0.3,
        labelspacing=0.1,
        borderaxespad=0.5
    )
    ax.add_artist(legend1)

    legend2 = ax.legend(
        handles=[
            Line2D([0], [0],
                   marker='o', markersize=12,
                   markerfacecolor=colors[0],
                   markeredgecolor='black', markeredgewidth=2,
                   linestyle=''),
            Line2D([0], [0],
                   marker='o', markersize=12,
                   markerfacecolor=colors[1],
                   markeredgecolor='black', markeredgewidth=2,
                   linestyle=''),
            Line2D([0], [0],
                   marker='o', markersize=12,
                   markerfacecolor=colors[2],
                   markeredgecolor='black', markeredgewidth=2,
                   linestyle='')
        ],
        labels=labelz,
        loc='lower right',
        bbox_to_anchor=(1, -0.12),
        fontsize=26,
        frameon=False,
        handlelength=1.5,
        handletextpad=0.3,
        labelspacing=0.1,
        borderaxespad=0.5
    )

    plt.tight_layout()
    fig.savefig("2_1_pareto1.pdf", format="pdf", bbox_inches="tight", pad_inches=0)
    plt.show()


# ----------------------------------------------------------------
# --- Main -------------------------------------------------------
# ----------------------------------------------------------------

if __name__ == "__main__":
    perf_constant, perf_spiky = gather_throughputs("figure_4")
    plot_figure3(perf_constant, perf_spiky)
```

```python
### Figures 6 ###

#!/usr/bin/env python3
import json
import glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Your 2–3-letter mnemonic names
PATTERN_NAMES = ["1-CON", "2-BUR", "3-SPI", "4-CON_O", "5-REG"]

def load_start_times_in_minutes(file_path):
    with open(file_path, 'r') as f:
        data = json.load(f)
    return pd.Series(q["start"] / 60000.0 for q in data.get("queries", []) if "start" in q)

def main():
    # Pick up to 5 pattern files
    json_files = sorted(glob.glob("figure_6/query_stream_*.json"))[:5]
    if not json_files:
        print("No JSON files found (query_stream_*.json).")
        return

    # ------------------------------------------------------------------
    # Styling: bump all text to ~17.5pt, thicker lines
    plt.style.use('default')
    plt.rcParams.update({
        'figure.figsize': (8, 2.5),
        'figure.dpi': 100,
        'savefig.dpi': 300,

        'font.size': 17.5,          # base font size
        'axes.titlesize': 17.5,
        'axes.labelsize': 17.5,
        'xtick.labelsize': 17.5,
        'ytick.labelsize': 17.5,
        'legend.fontsize': 17.5,

        'lines.linewidth': 3,       # thicker plot lines
        'lines.markersize': 7,
    })
    # ------------------------------------------------------------------

    fig, ax = plt.subplots(figsize=(8,2.5))

    # 20 bins from 0 to 1500 minutes
    bins = np.linspace(0, 1500, 21)
    cmap = plt.cm.get_cmap('Set1', len(json_files))
    linestyles = ['-', '--', '-.', ':', (0, (3,1,1,1))]
    markers    = ['o', 's', '^', 'D', 'X']

    for idx, file_path in enumerate(json_files):
        starts = load_start_times_in_minutes(file_path)
        counts, edges = np.histogram(starts, bins=bins)
        centers = 0.5 * (edges[:-1] + edges[1:])
        ax.plot(
            centers, counts,
            color=cmap(idx),
            linestyle=linestyles[idx],
            marker=markers[idx],
            label=PATTERN_NAMES[idx]
        )

    ax.set_xlabel("Time (minutes)")
    ax.set_ylabel("Query\nCount")

    # Dashed grid behind data
    ax.grid(True, linestyle='--', linewidth=1.0, alpha=1.0)
    ax.set_axisbelow(True)

    # L-shaped spines
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    for side in ['bottom','left']:
        ax.spines[side].set_color('black')
        ax.spines[side].set_linewidth(3)

    # Horizontal legend on top, centered
    ax.legend(
        title="Pattern",
        loc='upper left',
        bbox_to_anchor=(0,1),
        ncol=3,
        frameon=False,
        handlelength=1.5,
        handletextpad=0.3,
        columnspacing=0,
        labelspacing=0.2,
        borderpad=0.2,
        borderaxespad=0.2
    )

    plt.tight_layout()
    output_file = "4_standalone_patterns.pdf"
    plt.savefig(output_file)
    plt.show()
    plt.close(fig)
    print(f"Saved line-distribution plot to {output_file}")

if __name__ == "__main__":
    main()
```

```python
# Figures 7 and 8 #
# To note: CPU Time for 2.sql from Snowflake - 2.sql value from 2 nodes to 4 nodes
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
# restore our “talk” context + big fonts
sns.set_theme(style='whitegrid', context='talk', font_scale=1.8)


plt.rcParams.update({
    'axes.labelsize': 20,
    'axes.titlesize': 20,
    'xtick.labelsize': 20,
    'ytick.labelsize': 20,
    'legend.fontsize': 20,
    'font.size': 20,
    'lines.linewidth': 3,
    'lines.markersize': 12,
})
# at the top, alongside your other colour constants
LEGEND_COLORS = {
    'CPU':   '#5e4fa2',  # light rose
    'I/O':   '#21918c',  # pale blue
    'Mixed': '#fde725',  # mint green
}
LEGEND_COLORS2 = {
    'Scan':   '#5e4fa2',  # light rose
    'Agg':   '#21918c',  # pale blue
    'Join': '#fde725',  # mint green
}

LEGEND_COLORS3 = {
    'a': '#5e4fa2',
    'b': '#21918c'
}
MODEL_PALETTE = [
    "##5e4fa2",  # teal
    "#3b528b",  # burnt orange
    "#7570b3",  # purple
    "#e7298a",  # magenta
    "#66a61e",  # green
]



# Read CSVs
files_th = ['figure_7_and_8/sheet_1.csv', 'figure_7_and_8/sheet_2.csv', 'figure_7_and_8/sheet_3.csv']
files_ds = ['figure_7_and_8/sheet_4.csv', 'figure_7_and_8/sheet_5.csv', 'figure_7_and_8/sheet_6.csv']
dfs_th = [pd.read_csv(f) for f in files_th]
dfs_ds = [pd.read_csv(f) for f in files_ds]
combined = pd.concat(dfs_th + dfs_ds, ignore_index=True)
df_th = pd.concat(dfs_th, ignore_index=True)
df_ds = pd.concat(dfs_ds, ignore_index=True)

# # Color constants
# COLOR_TH  = '#D62728'  # red
# COLOR_DS  = '#1F77B4'  # blue
# COLOR_CPU = COLOR_TH
# COLOR_IO  = '#2CA02C'
# COLOR_MIX = COLOR_DS

# Helper to concat and drop zeros/nulls
def concat_nonzero(dfs, col):
    df = pd.concat(dfs, ignore_index=True)
    return df[col].dropna().loc[lambda x: x != 0].values

# Precompute all of your distributions...
qerror_th = [concat_nonzero(dfs_th, c) for c in
    ['summed','cpu_only','max_offset','powermean_offset','multiwave_offset']]
qerror_ds = [concat_nonzero(dfs_ds, c) for c in
    ['summed','cpu_only','max_offset','powermean_offset','multiwave_offset']]

scale_th_max   = [concat_nonzero([dfs_th[i]], 'max_offset') for i in range(3)]
scale_th_power = [concat_nonzero([dfs_th[i]], 'powermean_offset') for i in range(3)]

sf_data = [
    combined.loc[(combined['total_cpu_sec'] > 1)&(combined['size']==s), 'powermean_offset'].values
    for s in range(15,20)
]

phase_data = [
    df_th.loc[df_th['total_cpu_sec']>1, 'io_acc'].values,
    df_th.loc[df_th['total_cpu_sec']>1, 'shuffle_acc'].values,
    df_ds.loc[df_ds['total_cpu_sec']>1, 'io_acc'].values,
    df_ds.loc[df_ds['total_cpu_sec']>1, 'shuffle_acc'].values,
]

dom_data = [
    combined.loc[(combined['total_cpu_sec']>1)&(combined['type']==i), 'powermean_offset'].values
    for i in (1,2,3)
]

latency_data = [
    combined.loc[(combined['total_cpu_sec']>1)&(combined['short']==1), 'max_offset'].values,
    combined.loc[(combined['total_cpu_sec']>1)&(combined['short']==0)&
                  (combined['exec_wallclock_s']>=1)&(combined['exec_wallclock_s']<5),
                  'powermean_offset'].values,
    combined.loc[(combined['total_cpu_sec']>1)&(combined['exec_wallclock_s']>=5)&
                  (combined['exec_wallclock_s']<10), 'powermean_offset'].values,
    combined.loc[(combined['total_cpu_sec']>1)&(combined['exec_wallclock_s']>=10)&
                  (combined['exec_wallclock_s']<20), 'powermean_offset'].values,
    combined.loc[(combined['total_cpu_sec']>1)&(combined['long']==1), 'powermean_offset'].values,
]

# filter for “long” queries
df_long = combined[combined['exec_wallclock_s']>=1]
scan_data = [
    df_long[df_long['total_scan_mb']<=1000]['powermean_offset'].values,
    df_long[(df_long['total_scan_mb']>1000)&(df_long['total_scan_mb']<=5000)]['powermean_offset'].values,
    df_long[df_long['total_scan_mb']>5000]['powermean_offset'].values,
]
agg_data = [
    df_long[df_long['aggs']==1]['powermean_offset'].values,
    df_long[(df_long['aggs']>=2)&(df_long['aggs']<=3)]['powermean_offset'].values,
    df_long[df_long['aggs']>=4]['powermean_offset'].values,
]
join_data = [
    df_long[df_long['joins']<=3]['powermean_offset'].values,
    df_long[(df_long['joins']>3)&(df_long['joins']<10)]['powermean_offset'].values,
    df_long[df_long['joins']>=10]['powermean_offset'].values,
]
scan_data_mw = [
    df_long[df_long['total_scan_mb']<=1000]['multiwave_offset'].values,
    df_long[(df_long['total_scan_mb']>1000)&(df_long['total_scan_mb']<=5000)]['multiwave_offset'].values,
    df_long[df_long['total_scan_mb']>5000]['multiwave_offset'].values,
]
agg_data_mw = [
    df_long[df_long['aggs']==1]['multiwave_offset'].values,
    df_long[(df_long['aggs']>=2)&(df_long['aggs']<=3)]['multiwave_offset'].values,
    df_long[df_long['aggs']>=4]['multiwave_offset'].values,
]
join_data_mw = [
    df_long[df_long['joins']<=3]['multiwave_offset'].values,
    df_long[(df_long['joins']>3)&(df_long['joins']<10)]['multiwave_offset'].values,
    df_long[df_long['joins']>=10]['multiwave_offset'].values,
]

shuffle_data = [
    combined[(combined['shuffle_mb']<=1000)]['powermean_offset'].values,
    combined[(combined['shuffle_mb']>1000)&(combined['shuffle_mb']<=5000)]['powermean_offset'].values,
    combined[(combined['shuffle_mb']>5000)]['powermean_offset'].values,
]

# categories
models      = ['A','B','C','D','E']
nodes       = ['2','4','8']
sf_cats     = ['SF50','SF85','SF125','SF200','SF350']
latency_cats= ['<1s','1-5s','5-10s','10-20s','>20s']
shuffle_cats= ['Small\n(0–1 GB)','Medium\n(1–5 GB)','Large\n(>5 GB)']
comp_cats   = ['Scan','Agg','Join']
dom_resources = ['CPU','I/O','Mixed']

plt.rcParams.update({'font.size':20})
CELL_SIZE=(5,2.5)

# Colors and styling
COLOR_TH = '#D62728'
COLOR_DS = '#1F77B4'
PANEL_COLS = ['#2CA02C','#FF7F0E','#9467BD','#8C564B','#E377C2','#17BECF']
# Instead of red/blue/green, use:
PANEL_COLS # Bright contrasting colour palette
VIRIDIS_PALETTE = [
    "#5e4fa2",  # rgb(68, 1, 84)
    "#3b528b",  # rgb(59, 82, 139)
    "#21918c",  # rgb(33, 145, 140)
    "#5ec962",  # rgb(94, 201, 98)
    "#fde725",  # rgb(253, 231, 37)
]
VIRIDIS_PALETTE2 = [
    "#5e4fa2",  # rgb(68, 1, 84)
    "#21918c",  # rgb(33, 145, 140)
    "#fde725",  # rgb(253, 231, 37)
]

# Example assignments
COLOR_TH = VIRIDIS_PALETTE[0]
COLOR_DS = VIRIDIS_PALETTE[2]
PANEL_COLS = VIRIDIS_PALETTE

# Or, for exactly five model colors
# MODEL_COLOURFUL_PALETTE = COLOURFUL_PALETTE

CELL_SIZE = (5, 2.5)
plt.rcParams.update({'font.size': 20})

def draw_violin(ax, data, pos, color):
    vp = ax.violinplot(data, [pos], widths=0.7, showextrema=False)
    for b in vp['bodies']:
        b.set_facecolor(color)
        b.set_edgecolor('black')
        b.set_linewidth(1.5)
        b.set_alpha(1.0)             # <<< force full opacity
    q25, q50, q75 = np.percentile(data, [25, 50, 75])
    ax.plot([pos-0.2, pos+0.2], [q50, q50], color='black', linewidth=2.5)
    ax.plot([pos-0.15, pos+0.15], [q25, q25], color='black', linewidth=1.5)
    ax.plot([pos-0.15, pos+0.15], [q75, q75], color='black', linewidth=1.5)
    return q50, pos


for idx in range(8):
    fig, ax = plt.subplots(figsize=CELL_SIZE)

    if idx == 0:
        # Panel 1
        for i in range(len(qerror_th)):
            base = i*2 + 1
            draw_violin(ax, qerror_th[i], base, COLOR_TH)
            draw_violin(ax, qerror_ds[i], base+0.8, COLOR_DS)
        ax.set_xticks([i*2+1.4 for i in range(len(qerror_th))])
        ax.set_xticklabels(['Sum','CPU','Max','PM','MW'])
        handles = [
            Patch(facecolor=LEGEND_COLORS3['a'], edgecolor='black', label='TPC-H'),
            Patch(facecolor=LEGEND_COLORS3['b'], edgecolor='black', label='TPC-DS'),
        ]
        labels = ['TPC-H', 'TPC-DS']
        ax.legend(
            handles=handles,
            loc='upper center',
            bbox_to_anchor=(0.5, 1.10),
            ncol=2,
            frameon=False,
            handletextpad=0.3,   # space between patch and text
            columnspacing=0.5,   # space between columns
        )
        ax.set_ylabel('QERROR')

    elif idx == 1:
        # Panel 2
        re_cols = ['summed_re','cpu_only_re','max_offset_re','powermean_offset_re','multiwave_offset_re']
        rel_th = [concat_nonzero(dfs_th, c) for c in re_cols]
        rel_ds = [concat_nonzero(dfs_ds, c) for c in re_cols]
        all_vals = []
        for i in range(len(re_cols)):
            base = i*2 + 1
            draw_violin(ax, rel_th[i], base, COLOR_TH)
            draw_violin(ax, rel_ds[i], base+0.8, COLOR_DS)
            all_vals += list(rel_th[i]) + list(rel_ds[i])
        ax.set_xticks([i*2+1.4 for i in range(len(re_cols))])
        ax.set_xticklabels(['Sum','CPU','Max','PM','MW'])
        yt = sorted(set(list(ax.get_yticks()) + [-1,1]))
        ax.set_yticks(yt)
        handles = [
            Patch(facecolor=LEGEND_COLORS3['a'], edgecolor='black', label='TPC-H'),
            Patch(facecolor=LEGEND_COLORS3['b'], edgecolor='black', label='TPC-DS'),
        ]
        labels = ['TPC-H', 'TPC-DS']
        ax.legend(
            handles=handles,
            loc='upper center',
            bbox_to_anchor=(0.5, 1.10),
            ncol=2,
            frameon=False,
            handletextpad=0.3,   # space between patch and text
            columnspacing=0.5,   # space between columns
        )
        ax.set_ylabel('QERROR')
        lo, hi = min(all_vals), max(all_vals)
        pad = 0.1*(hi-lo)
        ax.set_ylim(lo-pad, hi+pad)
        ax.set_ylabel('Signed RelError')

    elif idx == 2:
        # Panel 3
        for i in range(len(scale_th_max)):
            base = i*2 + 1
            draw_violin(ax, scale_th_max[i], base, COLOR_TH)
            draw_violin(ax, scale_th_power[i], base+0.8, COLOR_DS)
        ax.set_xticks([i*2+1.4 for i in range(len(scale_th_max))])
        ax.set_xticklabels(['2','4','8'])
        ax.set_xlabel('Number of nodes')
        ax.set_ylabel('QERROR')
        handles = [
            Patch(facecolor=LEGEND_COLORS3['a'], edgecolor='black', label='max'),
            Patch(facecolor=LEGEND_COLORS3['b'], edgecolor='black', label='pm'),
        ]
        ax.legend(
            handles=handles,
            loc='upper center',
            bbox_to_anchor=(0.5, 1.10),
            ncol=2,
            frameon=False,
            handletextpad=0.3,   # space between patch and text
            columnspacing=0.5,   # space between columns
        )

    elif idx == 3:
        # Panel 4
        for i,d in enumerate(sf_data):
            draw_violin(ax, d, i+1, PANEL_COLS[i])
        ax.set_xticks(range(1,6))
        ax.set_xticklabels(['50','85','125','200','350'])
        ax.set_ylabel('QERROR')

    elif idx == 4:
        # Panel 5
        for i,d in enumerate(latency_data):
            draw_violin(ax, d, i+1, PANEL_COLS[i])
        bins_lbl = ['<1', '1–5', '5–10', '10–20', '≥20']
        ax.set_xticks(range(1, 6))
        ax.set_xticklabels(bins_lbl, rotation=0, ha='center')
        ax.set_ylabel('QERROR')

    elif idx == 5:
        # Panel 6: Scan / Agg / Join violin plots
        for i, d in enumerate(scan_data):
            draw_violin(ax, d, i+1, VIRIDIS_PALETTE2[0])
        for i, d in enumerate(agg_data):
            draw_violin(ax, d, i+5, VIRIDIS_PALETTE2[1])
        for i, d in enumerate(join_data):
            draw_violin(ax, d, i+9, VIRIDIS_PALETTE2[2])

        # place ticks under each group at the “middle” positions
        ax.set_xticks([2, 6, 10])
        ax.set_xticklabels([
            '0–1–5GiB',  # your “min operators / data” for Scan
            '1–2–4',      # for Agg
            '1–3–10'      # for Join
        ], fontsize=18)

        ax.set_ylabel('QERROR', fontsize=20)
        handles = [
            Patch(facecolor=LEGEND_COLORS2[r],
                  edgecolor='black',
                  label=r)
            for r in ['Scan','Agg','Join']
        ]
        ax.legend(
            handles=handles,
            loc='upper center',
            bbox_to_anchor=(0.5, 1.10),
            ncol=3,
            frameon=False,
            handletextpad=0.3,
            columnspacing=0.5,
        )


    elif idx == 6:
        # Panel 7
        resources = ['CPU','I/O','Mixed']
        colors = ['#D62728','#2CA02C','#1F77B4']
        for pos, data, col in zip([1,2,3], dom_data, VIRIDIS_PALETTE2):
            draw_violin(ax, data, pos, col)
        ax.set_xticks([])
        ax.set_xlabel('Dominant Resource')
        handles = [Patch(facecolor=c, edgecolor='black', label=r) for c,r in zip(colors,resources)]
        # now build a clean legend with the exact same solid colours
        handles = [
            Patch(facecolor=LEGEND_COLORS[r],
                  edgecolor='black',
                  label=r)
            for r in resources
        ]
        ax.set_ylabel('QERROR')
        ax.legend(
            handles=handles,
            loc='upper center',
            bbox_to_anchor=(0.5, 1.10),
            ncol=3,
            frameon=False,
            handletextpad=0.3,   # space between patch and text
            columnspacing=0.5,   # space between columns
        )

    else:
        # Panel 8
        shuffle_cats = ['[0,1GB)','[1,5GB)','[5GB,∞)']
        for i in range(3):
            base = i*2 + 1
            draw_violin(ax, shuffle_data[i], base, VIRIDIS_PALETTE2[0])
            mw = df_long.loc[
                (df_long['shuffle_mb'] > (1000 if i else -np.inf)) &
                (df_long['shuffle_mb'] <= (5000 if i==1 else np.inf)),
                'multiwave_offset'
            ].dropna().values
            draw_violin(ax, mw, base+0.8, VIRIDIS_PALETTE2[1])
        ax.set_xticks([1.4,3.4,5.4])
        ax.set_xticklabels(shuffle_cats)
        ax.set_ylabel('QERROR')
        handles = [
            Patch(facecolor=LEGEND_COLORS2['Scan'], edgecolor='black', label='pm'),
            Patch(facecolor=LEGEND_COLORS2['Agg'], edgecolor='black', label='mw'),
        ]
        ax.legend(handles=handles, loc='upper right', frameon=False)

    ax = plt.gca()

    # Hide the top & right spines
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    # Make bottom & left spines black & thick
    ax.spines['bottom'].set_color('black')
    ax.spines['left'].set_color('black')
    ax.spines['bottom'].set_linewidth(3)
    ax.spines['left'].set_linewidth(3)
    if idx != 1:
        ax.set_ylim(1,4)
    plt.tight_layout(pad=0.2)
    fig.savefig(f'panel_{idx+1}.pdf', bbox_inches='tight')
    plt.show()
    plt.close(fig)
```
